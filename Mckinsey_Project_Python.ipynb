# Mckinsey Dataset Project

```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/DillliBabu/mckinsey/refs/heads/main/mckinsey.csv"
output_file = "mckinsey.csv"
df=pd.read_csv(url)
```

```python
df.head(5)
```

```python
#Filter all records where continent is Asia
df[df['continent']=='Asia'].head(11)
```

```python
# Filter all records for the year 2007.
df[df['year']==2007].head(11)
```

```python
# Find all recods  where life_exp is greater than 70.
df[df['life_exp']>70].head(11)
```

```python
# Find countries with gdp_cap greater than 10,000.
df[df['gdp_cap']>10000]['country']
```

```python
#Filter records where population is more than 100 million.
df[df['population'] > 100000000].head(11)
```

```python
#Sort the dataset by life_exp in descending order.
df['life_exp'].sort_values(ascending=False).head(11)
```

```python
#Sort the dataset by gdp_cap in ascending order.
df['gdp_cap'].sort_values(ascending=True).head(11)
```

```python
#Select only country, year, and life_exp columns.
df[['country','year','life_exp']].head(11)
```

```python
#Find all records for India.
df[df['country']=='India'].head()
```

```python
#Find records where continent is either Europe or Asia.
df[~ df['continent'].isin (['Europe','Asia'])].head(11)
```

```python
#Find the average life expectancy for each continent.
df.groupby('continent')['life_exp'].mean()
```

```python
#Find total population per continent.
df.groupby('continent')['population'].sum()
```

```python
#Find the maximum GDP per capita for each continent.
df.groupby('continent')['gdp_cap'].max()
```

```python
#Find the minimum life expectancy for each year.
df.groupby('life_exp')['year'].max()
```

```python
#Find average GDP per capita for each year.
df.groupby('gdp_cap')['year'].mean()
```

```python
#Find the country with highest population in each year.
df.loc[df.groupby('year')['population'].idxmax(),['country','population','year']]
```

```python
#Count number of records per continent.
df['continent'].value_counts()
```

```python
#Find top 5 countries with highest life expectancy in 2007.
df[df['year']==2007].sort_values(by='life_exp', ascending=False).head(5)[['country','life_exp']]
```

```python
#Find bottom 5 countries with lowest GDP per capita in 1952.
df[df['year']==1952].sort_values(by='gdp_cap', ascending=True).head(5)[['country','year','gdp_cap']]
```

```python
df[df['year']==1952].sort_values('life_exp').head(11)[['year','continent','gdp_cap']]
```

```python
#Find median life expectancy foreachcontinent.
df.groupby('continent')['life_exp'].median()
```
